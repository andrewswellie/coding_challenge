```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as stats
import statsmodels.api as sm
```

```python
# Import CSV File and show the first 5 rows
wine_data_csv = "CSV/winemag_data_pandata.csv"
wine_data = pd.read_csv(wine_data_csv)
wine_data.head()
```

```python
# determine how many initial records there are
len(wine_data)
```

```python
# Remove any null values from the data set
wine_data.dropna()
```

```python
# Count the reviews by country
country_count = wine_data.groupby(["country"]).count()["designation"]
country_count = pd.DataFrame(country_count)
country_count = country_count.reset_index()
country_count.head()
```

```python
# Select only instances where a country has more than 1000 reviews
country_count[country_count['designation'] > 1000]
```

```python
# Create a dataframe that shows the avg price by country
wine_price = wine_data.groupby(["country"]).mean()["price"]
wine_price = pd.DataFrame(wine_price)
wine_price = wine_price.reset_index()
wine_price.head()
```

```python
# Create a data frame that shows AVG points by country
wine_points = wine_data.groupby(["country"]).mean()["points"]
wine_points = pd.DataFrame(wine_points)
wine_points = wine_points.reset_index()
wine_points.head()
```

```python
#Concatenate those data frames into one for graphing
final_df = pd.concat([country_count, wine_price, wine_points],axis=1)
final_df.head()
```

```python
# Re-establish the parameter that a country had to have more than 1000 reviews
final_df[final_df['designation'] > 1000]
```

# Analysis

## generally there seems to be a positive linear correlation between the average price of a bottle of wine and the points it receives from Wine Enthusiast. This trend seems to continue until just before the 60 price point at which the points start to decline. 

### Given more time with this data I would have applied both a linear and logrithmic model to help predict points from price. At first glance it seems that the logarithmic model would be a better fit that linear because of the shape of the data. Also, the R-squared value of .19 is low enough to make me think an ordinary least squares linear regression will not be the best fit for this data.

```python
# Plot the relationship between the avg price vs the points 
sns.lineplot(data=final_df, x='price',y='points',color='#73cae7', linewidth=2.5, label='price vs. points')

plt.xlabel("AVG Price")
plt.ylabel("AVG Points")
plt.legend=('Best')
plt.gcf().set_size_inches(14,7)
```

```python
final_df.describe()
```

```python
#fit model for statsmodels regression

final_df = final_df[final_df.replace([np.inf, -np.inf], np.nan).notnull().all(axis=1)]

X = final_df['price']
y = final_df["points"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```
